# 통계적 가설검정
- 모집단의 모수에 관하여 두가지 가설을 세우고, 표본으로부터 계산되는 통계량을 이용하여 어느 가설이 옳은지 판단하는 통계적 방법
- 검정통계량 vs 임곗값
- p값 vs 유의수준

=> 순서: 가설설정->유의수준 결정 -> 검정통계량, p값 계산 -> p값<유의수준 : 귀무가설 기각/ p값>유의수준 : 귀무가설 채택

```python
import numpy as np
import pandas as pd
from scipy import stats
%precision 3
```

```python
# 14개의 감자칩 무게 데이터
df = pd.read_csv('data/ch11_potato.csv')
sample = np.array(df['무게'])
sample, len(sample)
```

```python
# 표본평균, 표본분산
s_mean = sample.mean()
s_var = sample.var()
s_mean, s_var
```

```python
# 가설 = 모평균이 130g이다. (모집단이 정규분포를 따르고 모분산이 9임을 알고있다는 전제)
# 모집단은 N(130, 9)_ 표본집단이 모집단을 따르므로 표본평균은 N(130, 9/14)를 따름(14개의 표본)

# 표본
rv = stats.norm(130, np.sqrt(9/14))
# 평균, 분산
print(rv.mean(), rv.var(), '\n')
# 랜덤으로 sample 10개 추출
print(rv.rvs(10), '\n')
# 확률변수 (확률의 합이 5%인 확률변수 확인)
print(rv.isf(0.95), rv.ppf(0.05)) 
```

=> 128g 미만일 확률이 5%미만임

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()
xs = np.linspace(127,133,60)

ax1.plot(xs,rv.pdf(xs))
ax2.plot(xs,rv.cdf(xs),color='black')
ax1.vlines(rv.ppf(0.05), 0,0.5, color='gray',ls='--')
ax1.set_xlabel("x",fontsize=20)
ax1.set_ylabel('pdf',rotation=0,fontsize=20)
ax2.set_ylabel('cdf',rotation=0,fontsize=20)
ax1.set_ylim
```

- 검정통계량과 임계값 활용

```python
# 검정통계량 z :표본평균을 표준화
z = (s_mean - 130)/np.sqrt(9/14)
z
```

```python
# 임계값
rv = stats.norm()
rv.isf(0.95)
```

=> 검정통계량<임계값 ==> 귀무가설 기각

- p값과 유의수준 활용

```python
# 검정통계량으로 부터 p값 산출
rv.cdf(z)
```

=> p값이 0.05보다 작으므로 귀무가설 기각 (감자튀김의 평균 무게는 130g이 아니다)

=> 양측검정의 p값은 상단과 하단의 양쪽 면적을 고려해야 하므로 누적밀도 함수의 2배로 함

### 가설검정에서의 두 가지 오류
- 제1종 오류 : 귀무가설이 옳을 때 귀무가설을 기각하는 오류
  - '실제로 평균이 130g'인데도 평균은 130g 보다 작다라는 결론을 내리는 상황 
- 제2종 오류 : 대립가설이 옳을 때 귀무가설을 채택하는 오류
  - '모평균은 130g 보다 작다'인데도 모평균은 130g 보다 작다라는 결론을 얻을 수 없는 비율
- 통계적 가설검정에서는 제1종 오류는 제어할 수 있지만 제2종 오류는 제어할 수 없는 비대칭성이 있다.

```python
# 1종 오류 
rv = stats.norm(130, 3)

# 14개의 표본을 추출하여 가설검정을 수행하는 작업을 10000번 실행
# 제 1종 오률를 범하는 비율인 '평균이 130보다 작다'라는 결론을 내리는 비율을 계산
c = stats.norm().isf(0.95)
n_samples = 10000
cnt = 0
for _ in range(n_samples):
    sample_ = np.round(rv.rvs(14), 2)
    s_mean_ = np.mean(sample_)
    z = (s_mean_ - 130)/np.sqrt(9/14)
    # z보다 임계값 보다 작으면 count+1 (원래는 귀무가설이 기각되는 영역)
    if z<c:
        cnt+=1
cnt / n_samples
```

```python
# 1%로 줄였을 때
c = stats.norm().isf(0.99)
n_samples = 10000
cnt = 0
for _ in range(n_samples):
    sample_ = np.round(rv.rvs(14), 2)
    s_mean_ = np.mean(sample_)
    z = (s_mean_ - 130)/np.sqrt(9/14)
    # z보다 임계값 보다 작으면 count+1 (원래는 귀무가설이 기각되는 영역)
    if z<c:
        cnt+=1
cnt / n_samples
```

```python
# 제 2종 오류 (대립가설 기각)

# '모평균이 130g보다 작다' 인데 이 결론을 얻을 수 없는 상황
# 모평균이 128g이다라는 설정 모집단 N(128, 9)
# 14개의 표본을 추출하여 가설검정을 수행하는 작업을 10000번 실행

rv = stats.norm(128, 3)

# 하위 5% 이하 확률변수
c = stats.norm().isf(0.95)
n_samples = 10000
cnt = 0
for _ in range(n_samples):
    sample_ = np.round(rv.rvs(14), 2)
    s_mean_ = np.mean(sample_)
    # 표준화(검정통계량)
    z = (s_mean_ - 130)/np.sqrt(9/14)
    # 검정통계량>=인계값(귀무가설 채택)
    if z>=c :
        cnt +=1
cnt/n_samples
```

=> 2종 오류를 범하는 비율 (B) : 1-B = 검정력, 분석가가 제어 할 수 없음

## 기본적인 가설검정 - 1표본

### 정규분포의 모평균에 대한 검정 : 모분산을 알고 있는 경우

```python
# 모평균이 어떤 값이 아니라고 주장하기 위한 검정
# 귀무가설 : 모평균 == m
# 대립가설 : 모평균 != m

def pmean_test(sample, mean0, p_var, alpha=0.05):
    # 표본
    s_mean = np.mean(sample)
    n = len(sample)
    rv = stats.norm()
    # 신뢰구간의 영역인 확률변수 구간 구하기
    interval = rv.interval(1-alpha)
    
    # mean0=>귀무가설에서 주장하는 m
    z = (s_mean-mean0)/np.sqrt(p_var/n)
    if interval[0] <= z <= interval[1]:
        print('귀무가설 채택')
    else:
        print('귀무가설 기각')
    if z <0:
        p = rv.cdf(z)*2
    else:
        p = (1-rv.cdf(z))*2
    print(f'p값은 {p:3f}')
```

```python
sample = np.array(df['무게'])
pmean_test(sample, 130, 9)
```

### 정규분포의 모분산에 대한 검정

```python
# 모분산이 sigma가 아니라고 주장하는 검정

def pvar_test(sample, var0, alpha = 0.05):
    u_var = np.var(sample, ddof=1)
    n = len(sample)
    rv = stats.chi2(n-1)
    interval = rv.interval(1-alpha)
    
    # var0 : 가설에서 주장하는 sigma
    y = (n-1)*u_var/var0
    if interval[0]<= y <= interval[1]:
        print('귀무가설 채택')
    else:
        print('귀무가설 기각')
        
    if y<rv.isf(0.5):
        p = rv.cdf(y)*2
    else:
        p = (1-rv.cdf(y))*2
    print(f'p값은 {p:3f}')
```

```python
pvar_test(sample, 9)
```

```python
### 정규분포의 모평균에 대한 검정 : 모분산을 모르는 경우 _t분포를 활용

def pmean_test(sample, mean0, alpha = 0.05):
    s_mean = np.mean(sample)
    u_var = np.var(sample, ddof=1)
    n = len(sample)
    rv = stats.t(n-1)
    interval = rv.interval(1-alpha)
    
    t = (s_mean-mean0)/ np.sqrt(u_var/n)
    if interval[0]<= t <= interval[1]:
        print('귀무가설 채택')
    else:
        print('귀무가설 기각')
    
    if t < 0 :
        p = rv.cdf(t)*2
    else:
        p = (1-rv.pdf(t))*2
    print(f' p값은 {p:3f}')
```

```python
pmean_test(sample, 130)
```

```python
# 위를 더 간단히 ttest_1samp함수 구현 (t검정통계량,p값 반환)
t, p = stats.ttest_1samp(sample, 130)
t, p
```

### 카이제곱검정
- 독립성 검정이란 두 변수 'X,Y에 관해서 X와 Y가 독립이다.'라는 귀무가설과 'X,Y에 관해서 X와 Y가 독립이 아니다.'라는 대립가설에 의해 수행되는 검정
- 독립성 검정에 카이제곱분포가 사용되기 때문에 카이제곱검정이라고도 부른다.
- 광고A와 광고B간의 구입 비율에 유의한 차이가 있는지 확인하는 경우 사용할 수 있음.

```python
ad_df = pd.read_csv('data/ch11_ad.csv')
print(ad_df.head())
len(ad_df)
```

```python
# 교차집계표 작성(각각의 광고로 인해 어느 정도의 상품을 구입했는지)
ad_cross = pd.crosstab(ad_df['광고'], ad_df['구입'])
ad_cross
```

```python
# 상품을 구힙한 비율 추출
ad_cross['했다']/(ad_cross['하지 않았다']+ad_cross['했다'])
```

```python
# 상품을 구입한 사람의 합계, 구입하지 않은 사람의 합계, 광고A를 본 사람의 합계, 광고B를 본 사람의 합계
n_not, n_yes = ad_cross.sum()
n_adA, n_adB = ad_cross.sum(axis=1)
n_not, n_yes, n_adA, n_adB
```

=> 광고 상관없이 상품을 구입한 비율 : 10% (광고와 구입이 독립이여서 광고에 상품 구입 비율이 변화가 없으면, A광고를 보고 상품을 구입할 비율이 10%이상이라 기대할 수 있음 : 기대도수)

```python
# 실제 관측된 데이터: 관측도수
# 기대도수 계산 (여기선 10%)

n = len(ad_df)
ad_ef = pd.DataFrame({'했다' : [n_adA*n_yes/n,n_adB*n_yes/n],
                     '하지 않았다': [n_adA*n_not/n, n_adB*n_not/n]},
                    index = ['A', 'B'])
ad_ef
```

=> A광고 후 40명이 살거라 기대, B광고 후 60명이 구매 한다 기대

```python
# 기대도수와 관측도수의 차이를 측정 => 카이제곱검정

y = ((ad_cross - ad_ef)**2 / ad_ef).sum().sum()
y
```

=> y는 자유도가 1인 카이제곱분포를 근사적으로 따름

```python
rv = stats.chi2(1)
# p값 추출
1 - rv.cdf(y)
```

=> 유의수준 0.05보다 크므로 귀무가설 채택 ('광고 A와 B가 유의한 차이가 나지 않음')

```python
# contingency 함수 사용 (인수:고차집계표, 반환: 검정통계량, p값, 자유도, 기대도수)
chi2, p, dof, ef = stats.chi2_contingency(ad_cross, correction=False)
chi2, p, dof
```
